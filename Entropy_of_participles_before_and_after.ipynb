```python
def get_info(sent, keyword):
    """Given a parsed sentence in conll format, return every head
    with its children (showing the position in the sent and the word form).
    Also return the sentence."""
    try:
        p = parse_tree(sent)
    except: # could not parse, happens with a few weird sentences with no ROOT lke "uh...and"
        return [None, None]
    for curp in p:
      stack = [curp]
      a_nary = []
      words = []
      while len(stack) > 0:
          cur = stack.pop(0)
          curargs = []
          cur_list = [cur.token["id"], cur.token["form"], cur.token["upostag"]]
          nary_args = {"head": cur_list, "children": []}
          words += [(cur.token["id"], cur.token["form"])]
          for i in cur.children:
              if (i.token["deprel"] == "punct" or cur.token["deprel"] == "punct"):
                  continue
              child_list = [i.token["id"], i.token["deprel"], i.token["form"]]
              nary_args["children"] += [child_list]
              stack += [i]
          a_nary += [nary_args]
      if keyword in [i[1] for i in words]:
        return a_nary, " ".join([i[1] for i in sorted(words)])

def get_deps(doc, keyword):
  """Parse a sentence and return its info, as in get_info"""
  #doc = nlp(s)
  return get_info(doc._.conll_str, keyword)
```

```python
def filter_args(dep, keyword):
  result = []
  if dep["head"][1] == keyword:
    return True
  for word in dep["children"]:
    if keyword == word[-1]:
      return True
  return False


def _get_keywords(doc, keyword, construction):
  try:
    deps = get_deps(doc, keyword)[0]
  except TypeError:
    return ["", ""]
  #return deps # UNDO THIS
  results = {"word": keyword,
             "headnoun": [],
             "modnoun": []}
  preps_nums = []
  if deps is None or deps[0] is None:
    return ["", ""]
  for keydep in deps:
      for child in keydep["children"]:
        if "Prenominal" in construction or "passive name" in construction.lower():
          if child[1] in modcats and keydep["head"][1] == keyword:
            results["modnoun"] +=  [child[2]]
          if child[2] == keyword:
            results["headnoun"] += [keydep["head"][1]]
        elif construction == "Relative":
          if keydep["head"][1] == keyword:
            preps_nums += [child[0]]
          if child[2] == keyword:
            results["headnoun"] += [keydep["head"][1]]
        elif construction == "Passive":
          if keydep["head"][1] == keyword and child[1] in ["prep", "agent"] and keydep["head"][0] < child[0]:
            preps_nums += [child[0]]
          if keydep["head"][1] == keyword and child[1] == "nsubjpass":
            results["headnoun"] += [child[2]]
  if construction in ["Relative", "Passive"]:
    for keydep in deps:
      if keydep["head"][0] in preps_nums:
        for child in keydep["children"]:
          if child[1] == "pobj":
            results["modnoun"] += [child[2]]
  return [",".join(results["headnoun"]).split(",")[0], ",".join(results["modnoun"]).split(",")[0]]

def get_keywords(doc, keyword, construction, ind=-1):
  result = _get_keywords(doc, keyword, construction)
  #return result # UNDO THIS
  if result[0] == "" or result[1] == "":
    for cons in construction_types:
       tryresult = _get_keywords(doc, keyword, cons)
       if tryresult[0] != "" and tryresult[1] != "":
         return tryresult
  return result


# run this to just use all constructions
construction_types =  list(set(d.Construction))
```
